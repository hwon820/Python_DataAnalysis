##015. 금액의 결측치 수정

```python
#결측치 확인
uriage_data.isnull().any(axis = 0)  #변수별 결측치 유무. item_price에 결측치 존재 확인
```

```python
#결측치 처리

flg_is_null = uriage_data["item_price"].isnull()
for trg in list(uriage_data.loc[flg_is_null, "item_name"].unique()):
  #결측치가 아닌 데이터이고 현재 trg(상품 이름)과 같은 상품의 가격(item_price)를 가져와 price에 저장
  price = uriage_data.loc[(~flg_is_null) & (uriage_data["item_name"] == trg), "item_price"].max() #max는 왜?
  #결측치가 맞는 데이터의 item_price에 방금 저장한 price를 상품명에 맞춰 저장
  uriage_data["item_price"].loc[(flg_is_null) & (uriage_data["item_name"] == trg)] = price

uriage_data.head()
```

```python
#결측치 처리 후 확인 및 검증
uriage_data.isnull().any(axis = 0)
```

```python
for trg in list(uriage_data["item_name"].sort_values().unique()):
  print(trg + "의 최고가:" + str(uriage_data.loc[uriage_data["item_name"] == trg]["item_price"].max()) + "의 최저가:" +
        str(uriage_data.loc[uriage_data["item_name"] == trg]["item_price"].min(skipna = False)))
#skipna : Nan값 무시 여부 설정  
#최대금액과 최소금액 일치 -> 수정 성공
```

##017. 날짜 오류 수정

```python
#날짜 포맷 통일

#숫자로 읽히는 데이터 확인
flg_is_serial = kokyaku_data["등록일"].astype("str").str.isdigit()
flg_is_serial.sum()
#22개
```

```python
#숫자로 등록된 부분 수정

fromSerial = pd.to_timedelta(kokyaku_data.loc[flg_is_serial, "등록일"].astype("float"), unit="D") + pd.to_datetime("1900/01/01")
fromSerial

#pd.to_timedelta( ) : 숫자를 날짜로 변환. loc( )를 통해 조건에 맞는 데이터만 추출해서 변경 가능
```

```python
#서식 통일
fromString = pd.to_datetime(kokyaku_data.loc[~flg_is_serial, "등록일"])
fromString
```

```python
#결합
kokyaku_data["등록일"] = pd.concat([fromSerial, fromString])
kokyaku_data
```

```python
#집계

kokyaku_data["등록연월"] = kokyaku_data["등록일"].dt.strftime("%Y%m")
rslt = kokyaku_data.groupby("등록연월").count()["고객이름"]
print(rslt)
print(len(kokyaku_data))
```

```python
#숫자로 읽히는 데이터 재확인
flg_is_serial = kokyaku_data["등록일"].astype("str").str.isdigit()
flg_is_serial.sum()
#0개 -> 수정 성공
```

##018. 고객 이름을 key로 두 데이터 join

```python
#각 데이터의 key가 되는 column이름을 지정하고 join
join_data = pd.merge(uriage_data, kokyaku_data, left_on = "customer_name", right_on = "고객이름", how = "left")
#중복 column중 customer_name 삭제
join_data = join_data.drop("customer_name", axis = 1)
join_data
```

##019. 정제한 데이터 덤프

```python
#정제한 데이터를 csv파일로 덤프(출력)
dump_data = join_data[["purchase_date", "purchase_month", "item_name", "item_price", "고객이름", "지역", "등록일"]] #column 배치 조정
dump_data
```

```python
#덤프파일 생성
dump_data.to_csv("dump_data.csv", index = False)
```
